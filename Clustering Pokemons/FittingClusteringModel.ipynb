```python
import numpy as np
import pandas as pd
import statsmodels as statsmodels
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import joblib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

data = pd.read_csv('./processedData.csv')
print(data)
```

```python
fig = plt.figure()
km1 = KMeans(n_clusters=2, max_iter=1000, tol=1e-8, algorithm='full').fit(data)
print(km1.labels_)
joblib.dump(km1, './km1.pkl')
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Defense'], data['Attack'], data['Total'], c=km1.labels_, cmap='rainbow')
km2 = KMeans(n_clusters=4, max_iter=1000, tol=1e-8, algorithm='full').fit(data)
print(km2.labels_)
joblib.dump(km2, './km2.pkl')
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Defense'], data['Attack'], data['Total'], c=km2.labels_, cmap='rainbow')
km3 = KMeans(n_clusters=8, max_iter=1000, tol=1e-8, algorithm='full').fit(data)
print(km3.labels_)
joblib.dump(km3, './km3.pkl')
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Defense'], data['Attack'], data['Total'], c=km3.labels_, cmap='rainbow')
km4 = KMeans(n_clusters=16, max_iter=1000, tol=1e-8, algorithm='full').fit(data)
print(km4.labels_)
joblib.dump(km4, './km4.pkl')
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Defense'], data['Attack'], data['Total'], c=km4.labels_, cmap='rainbow')
km5 = KMeans(n_clusters=32, max_iter=1000, tol=1e-8, algorithm='full').fit(data)
print(km5.labels_)
joblib.dump(km5, './km5.pkl')
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Defense'], data['Attack'], data['Total'], c=km5.labels_, cmap='rainbow')
```

```python
hc1 = AgglomerativeClustering().fit(data)
print(hc1.labels_)
joblib.dump(hc1, './hc1.pkl')
hc2 = AgglomerativeClustering(n_clusters=4).fit(data)
print(hc2.labels_)
joblib.dump(hc2, './hc2.pkl')
hc3 = AgglomerativeClustering(n_clusters=8).fit(data)
print(hc3.labels_)
joblib.dump(hc3, './hc3.pkl')
hc4 = AgglomerativeClustering(n_clusters=16).fit(data)
print(hc4.labels_)
joblib.dump(hc4, './hc4.pkl')
hc5 = AgglomerativeClustering(n_clusters=32).fit(data)
print(hc5.labels_)
joblib.dump(hc5, './hc5.pkl')
```
